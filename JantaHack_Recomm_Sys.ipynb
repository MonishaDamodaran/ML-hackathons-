```
import pandas as pd
import numpy as np
import turicreate as tc
```

```
train_data=pd.read_csv('drive/My Drive/Recomendation_systems/train.csv')
test_data=pd.read_csv('drive/My Drive/Recomendation_systems/test.csv')
combine_set=pd.concat([train_data,test_data], ignore_index=True)
```

```
combine_set_tc=tc.SFrame(combine_set)
item_similarity_model=tc.item_similarity_recommender.create(combine_set_tc, user_id='user_id', item_id='challenge', target='challenge_sequence', similarity_type='cosine')
unique_list=np.unique(test_data['user_id'])
item_recommender=item_similarity_model.recommend(users=unique_list,k=3)
item_recommender.print_rows(num_rows=24)
```

```
test_df=item_recommender.to_dataframe()
sample_submission=pd.read_csv('drive/My Drive/Recomendation_systems/sample_submission_J0OjXLi_DDt3uQN.csv')
sample_submission['challenge']=np.array(test_df['challenge'])
sample_submission.to_csv('drive/My Drive/Recomendation_systems/main_test.csv', index=False)
```
